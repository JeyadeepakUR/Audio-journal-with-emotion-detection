```python
RECORD_AND_PROCESS = """
const sleep = time => new Promise(resolve => setTimeout(resolve, time));
let stream;
let recorder;
let audioChunks;

const record = async () => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true });
  recorder = new MediaRecorder(stream);
  audioChunks = [];

  recorder.ondataavailable = e => audioChunks.push(e.data);
  recorder.start();

  alert("🎙️ Recording started! Speak now...");
  await sleep(5000);  // record for 5 seconds
  recorder.stop();

  await sleep(1000);
  stream.getTracks().forEach(track => track.stop());

  const blob = new Blob(audioChunks);
  const reader = new FileReader();
  reader.readAsDataURL(blob);
  reader.onloadend = () => {
    const b64 = reader.result.split(',')[1];
    google.colab.kernel.invokeFunction('notebook.handleAudio', [b64], {});
  };
};

record();
"""
```

```python
def handle_audio(b64):
    print("⏳ Processing audio...")
    audio_bytes = base64.b64decode(b64)
    audio = AudioSegment.from_file(io.BytesIO(audio_bytes), format="webm")
    audio.export("recorded.wav", format="wav")

    # Load Whisper
    model = whisper.load_model("medium")  # try 'medium' for better multilingual
    result = model.transcribe("recorded.wav", language="en")  # or 'ta', 'te'
    text = result["text"]
    print("\n📝 Transcribed Text:\n", text)

    # Sentiment Analysis
    senti = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
    sentiment = senti(text)
    print("\n🧠 Sentiment Analysis:\n", sentiment)

output.register_callback("notebook.handleAudio", handle_audio)
```

```python
display(Javascript(RECORD_AND_PROCESS))
```

```python
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer

# Use multilingual sentiment model as base (you can replace with OCEAN-style model later)
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Sample user journal input
user_journal = """
Lately, I’ve been questioning a lot of my decisions. I feel anxious and uncertain, but also hopeful sometimes.
I want to grow, but I'm afraid of making mistakes. Still, I know change is part of the process.
"""

# Sentiment analysis (simulated personality snapshot)
sentiment = sentiment_pipeline(user_journal)

# Personality inference mockup based on sentiment result
rating = sentiment[0]['label']
if '1' in rating or '2' in rating:
    profile = {
        "Neuroticism": "High",
        "Openness": "Moderate",
        "Conscientiousness": "Low",
        "Extraversion": "Low",
        "Agreeableness": "Moderate"
    }
elif '3' in rating:
    profile = {
        "Neuroticism": "Moderate",
        "Openness": "Moderate",
        "Conscientiousness": "Moderate",
        "Extraversion": "Moderate",
        "Agreeableness": "High"
    }
else:
    profile = {
        "Neuroticism": "Low",
        "Openness": "High",
        "Conscientiousness": "High",
        "Extraversion": "High",
        "Agreeableness": "High"
    }

print("\n🧠 Sentiment Result:", sentiment)
print("\n📊 Inferred Psychometric Profile (OCEAN-style):")
for trait, value in profile.items():
    print(f"{trait}: {value}")
```
